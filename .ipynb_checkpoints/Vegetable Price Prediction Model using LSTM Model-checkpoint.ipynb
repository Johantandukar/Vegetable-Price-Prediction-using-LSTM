# Vegetable Price Prediction Model using LSTM Model

## Load libraries

```python
# Used for interacting with the operating system
import os

# Used for arrays, mathematical operations, and data manipulation
import numpy as np

# Library for data analysis and manipulation
import pandas as pd

# Visualization library for plotting graphs
import matplotlib.pyplot as plt

# TensorFlow main library for building deep learning models
import tensorflow as tf

# Dense: Fully connected neural network layer
# LSTM: Long Short-Term Memory layer for sequence learning
# Bidirectional: Allows LSTM to process data forward and backward
# Lambda: Used to apply custom operations on tensors
from tensorflow.keras.layers import Dense, LSTM, Bidirectional, Lambda

# Sequential model allows stacking layers linearly
from tensorflow.keras.models import Sequential

# Huber loss function is robust to outliers which is commonly used in regression and time-series forecasting
from tensorflow.keras.losses import Huber

# Adam optimizer adapts learning rate for faster and stable convergence
from tensorflow.keras.optimizers import Adam

# Callback to dynamically change learning rate during training
from tensorflow.keras.callbacks import LearningRateScheduler

# Mean Absolute Error metric to evaluate model prediction accuracy
from tensorflow.keras.metrics import mean_absolute_error
```

## Read Data

```python
# Full path to the input file
FILE_IN = os.path.join("kalimati_tarkari_dataset_merged_cleaned.csv")
# Name of the fruit of which the price will be predicted
FRUIT_TYPE = "cauli local"
# Date used to split the training and the test part
SPLIT_DATE = "2023-01-01"
# Size of the window applied to the time series
WINDOW_SIZE = 10
# Batch size for model training and prediction
BATCH_SIZE = 32
# Buffer size used to shuffle the training set, should be bigger than the dataset's size
SHUFFLE_BUFFER = 3000
```

```python
# FILE_IN contains the path or filename of the dataset
# sep="," specifies that the file is comma-separated
df = pd.read_csv(FILE_IN, sep=",")

# Display the first 5 rows of the dataset
df.head()
```

## Data Overview

```python
# Filter the dataset to include only rows
df = df[df['name_original'] == FRUIT_TYPE]

# Display concise information about the filtered DataFrame
df.info()
```

```python
# Select only the required columns from the dataset (i.e., 'date' and  'average')
df = df[['date', 'average']]

df.head()
```

The date column was first converted into a datetime format to ensure that the dataset correctly represents temporal information and allows proper time-based operations. After conversion, the date column was set as the index of the DataFrame, replacing the default numerical index. This step is essential in time-series analysis as it enables efficient indexing, sorting, resampling, and visualization based on time. As a result, the original integer-based index (0, 1, 2, …) is removed, and the DataFrame is now organized chronologically using the date values.

```python
# Convert the 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'])

# Set the 'date' column as the DataFrame index
df.set_index('date', inplace=True)
df.head()
```

The code first converts the date column into a datetime format using pd.to_datetime(), ensuring that the dataset recognizes the entries as proper dates rather than strings.

Next, we will resample the series into dates, this will also ensure that there will be no duplicated dates after the resampling

```python
# Identify rows where the 'average' column has missing values (NaN)
missing_values = df[df['average'].isna()]

# Print the rows with missing values to inspect them
print(missing_values)
```

## Data Modelling

```python
def plot_series(x, y, start=0, end=None, title=None, xlabel=None, ylabel=None):
    """
    Utility function to plot a time series or multiple series.
    
    Args:
        x (1d Numpy array): values for the x-axis (e.g., dates or time steps)
        y (tuple of Numpy arrays or 1d Numpy array): values for the y-axis (e.g., prices)
        start (int): starting index of the data to plot
        end (int): ending index of the data to plot
        title (str): title of the plot
        xlabel (str): label for the x-axis
        ylabel (str): label for the y-axis
    """
    
    # Set the size of the figure to make it larger and easier to read
    plt.figure(figsize=(14, 8))
    
    # If y is a tuple, it means multiple series are provided
    if type(y) is tuple:
        # Plot each series individually
        for val_set in y:
            plt.plot(x[start:end], val_set[start:end])
    else:
        # Plot a single series
        plt.plot(x[start:end], y[start:end])

    # Add title and labels if provided
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    
    # Enable grid for better readability of values
    plt.grid(True)
    plt.show()
```

## Train-test timestep data creation

```python
# Split the dataset into training and testing sets based on SPLIT_DATE
df_train = df[df.index < SPLIT_DATE]
# All rows with index (date) on or after SPLIT_DATE go to testing
df_test = df[df.index >= SPLIT_DATE]

# Store the test set index (dates) separately for plotting predictions later
time_test = df_test.index

# training data
X_train = df_train['average'].values  
# testing data
X_test = df_test['average'].values    

# Print shapes of training and testing arrays to verify
print(f"X_train shape = {X_train.shape}; X_test shape = {X_test.shape}")

# Print first 10 values of training data to inspect
print(f"Some X_train data: {X_train[:10]}")
```

In time series forecasting, an effective approach is to use a limited number of past observations to predict future values. For instance, with a sliding window of size 10, the values from time steps t-1, t-2, …, t-10 are used to forecast the value at time t. Additionally, the TensorFlow Dataset class can be leveraged to transform time series data into a format that is fully compatible with TensorFlow. Using Dataset not only simplifies data handling but also enhances performance by enabling parallelized batch processing during model training.

```python
def window_train(series, window_size, batch_size, shuffle_buffer):
    """
    Convert the training series into a TensorFlow Dataset suitable for LSTM training.
    
    Args:
        series (1d Numpy array): input time series data
        window_size (int): number of past observations to use for predicting the next value
        batch_size (int): number of samples per batch
        shuffle_buffer (int): buffer size for shuffling the dataset
        
    Returns:
        TensorFlow Dataset where each element is a tuple:
        ([past values], [next value])
    """
    # Convert the input series into a TensorFlow Dataset
    dataset = tf.data.Dataset.from_tensor_slices(series)

    # Create windows of size (window_size + 1)
    # The last element in each window will be the label (target)
    dataset = dataset.window(window_size + 1, shift=1, stride=1, drop_remainder=True)
    
    # Flatten each window into a single batch of values
    dataset = dataset.flat_map(lambda w: w.batch(window_size + 1))
    
    # Split each window into input (past values) and label (next value)
    dataset = dataset.map(lambda w: (w[:-1], w[-1]))
    
    # Shuffle the dataset to prevent sequential order from biasing the training
    dataset = dataset.shuffle(shuffle_buffer)
    
    # Group into batches and prefetch to improve training performance
    dataset = dataset.batch(batch_size).prefetch(1)
    
    return dataset


def window_test(series, window_size, batch_size):
    """
    Convert the test series into a TensorFlow Dataset for evaluation.
    
    Args:
        series (1d Numpy array): input test series
        window_size (int): number of past observations used to predict the next value
        batch_size (int): number of samples per batch
        
    Returns:
        TensorFlow Dataset ready for testing
    """
    dataset = tf.data.Dataset.from_tensor_slices(series)
    
    # Use window_size (not +1) because labels are not separated in test set
    dataset = dataset.window(window_size, shift=1, stride=1, drop_remainder=True)
    
    # Flatten each window into a batch
    dataset = dataset.flat_map(lambda w: w.batch(window_size))
    
    # Group into batches and prefetch for efficiency
    dataset = dataset.batch(batch_size).prefetch(1)
    
    return dataset


# Create the training dataset from X_train
ds_train = window_train(X_train, WINDOW_SIZE, BATCH_SIZE, SHUFFLE_BUFFER)

# Create the test dataset
# Concatenate last WINDOW_SIZE elements of training set with test set (except the last value)
# This ensures we have enough past values to predict the first test value
ds_test = window_test(
    np.concatenate((X_train[-WINDOW_SIZE:], X_test[:-1])),
    WINDOW_SIZE,
    BATCH_SIZE
)

# Check the shapes of the first batch in training set
for batch in ds_train.take(1):
    print(batch[0].shape)  # Shape of input features
    print(batch[1].shape)  # Shape of labels

print("--------------------")

# Check the shape of the first batch in test set
for batch in ds_test.take(1):
    print(batch.shape)
```

Next, we will build the prediction model using an LSTM. TensorFlow’s Lambda layer can be utilized to simplify certain preprocessing steps. In this model, the first Lambda layer adds an extra dimension to the input because each batch is initially a 2D array ([batch_size, values]), whereas LSTM (and other RNN) layers require 3D inputs ([batch_size, timesteps, features]). The final Lambda layer is used to rescale the outputs. Since LSTM outputs typically fall within the range [-1, 1] due to the tanh activation, this scaling adjusts the predictions to match the actual price range (approximately 0 to 100), which can also help improve training efficiency.

## Forecasting

```python
# Clear any previous TensorFlow/Keras session which This helps avoid clutter and conflicts when creating a new model
tf.keras.backend.clear_session()

# Build a LSTM model
model = Sequential([

    # Expand input dimensions to make data compatible with LSTM layers
    # Converts shape from (WINDOW_SIZE,) to (WINDOW_SIZE, 1)
    Lambda(lambda x: tf.expand_dims(x, axis=-1), input_shape=[WINDOW_SIZE]),

    # First Bidirectional LSTM layer with 32 units
    # return_sequences=True allows stacking another LSTM layer
    Bidirectional(LSTM(32, return_sequences=True)),

    # Second Bidirectional LSTM layer with 32 units
    # Outputs the final hidden state
    Bidirectional(LSTM(32)),

    # Dense output layer to predict the next time step value
    Dense(1),

    # Scale the output to match the original data scale
    Lambda(lambda x: x * 100.0)
])

# Save the initial model weights
# This allows reuse of the same weights when tuning learning rates
init_weights = model.get_weights()

# Display the model architecture summary
model.summary()
```

In this work we will use Adam as the optimizer for the model training. However, choosing an appropriate learning is not as easy, and a good way to do that is to employ the LearningRateScheduler class to update the learning rate during training. Then, we plot the impact of different learning rates on the losses in order to pick the most appropriate one.

```python
# Define a learning rate scheduler that gradually increases the learning rate
# Starting from a very small value (1e-8) and increasing exponentially each epoch
lr_schd = LearningRateScheduler(lambda epoch: 1e-8 * 10**(epoch / 20))

# Compile the model using Huber loss (robust to outliers)
# and Adam optimizer for efficient gradient-based optimization
model.compile(loss=Huber(), optimizer=Adam())

# Train the model for 100 epochs on the training dataset
# while dynamically adjusting the learning rate using the scheduler
hist = model.fit(ds_train, epochs=100, callbacks=[lr_schd])
```

Now plot the losses along with the different learning rates

```python
# Create an array of learning rates ranging from 1e-8 to ~1e-3
# This is done exponentially to test how the model loss behaves at different learning rates
lr_arr = 1e-8 * (10 ** (np.arange(100) / 20))

# Set up the plot figure size
plt.figure(figsize=(10, 6))

# Enable grid for better readability
plt.grid(True)

# Plot the loss history against the learning rates using a logarithmic scale on x-axis
# semilogx is used because learning rates vary exponentially
plt.semilogx(lr_arr, hist.history['loss'])

# Customize tick parameters for better visual appearance
plt.tick_params('both', length=10, width=1, which='both')

# Set the axes limits
# x-axis: from 1e-8 to 1e-3 (learning rate range)
# y-axis: from 0 to 100 (loss values)
plt.axis([1e-8, 1e-3, 0, 100])
```

As a general guideline, the learning rate should be selected at a value where the loss appears stable. In this example, a learning rate of 1e-4 is suitable. After determining this, the model is recompiled, and its weights are reset using the previously saved init_weights variable. This allows training to be run for a significantly larger number of epochs, enabling the model to better learn the patterns in the data.

```python
# Reset the model's weights to a previously saved state
model.set_weights(init_weights)

# Compile the model
# - Loss: Huber loss (robust to outliers)
# - Optimizer: Adam with a learning rate of 0.0001
# - Metrics: Track Mean Absolute Error (MAE) during training
model.compile(loss=Huber(), optimizer=Adam(learning_rate=1e-4), metrics=['mae'])

# Train the model on the training dataset for 300 epochs
# The training history (loss and metrics per epoch) is stored in 'hist'
hist = model.fit(ds_train, epochs=300)
```

Next, we visualize the model’s training progress by plotting both the loss and the Mean Absolute Error (MAE) against the number of epochs. This allows us to assess how the model improves over time and to identify trends such as convergence or overfitting.

```python
# Extract Mean Absolute Error (MAE) values from training history
hist_mae = hist.history['mae']

# Extract loss values from training history
hist_loss = hist.history['loss']

# Plot both MAE and loss over epochs
# - X-axis: epoch numbers (0 to total epochs-1) and Y-axis: MAE and loss values
plot_series(
    range(len(hist_loss)), (hist_mae, hist_loss), title="MAE & Loss", xlabel="Epochs"
)
```

Although both the loss and MAE continue to decrease slightly, it is reasonable to stop training at this stage and move on to making predictions on the test set. After obtaining the predictions, we plot them alongside the actual prices to visually evaluate the model’s performance.

## Prediction Analysis

```python
# Use the trained model to make predictions on the test dataset
forecasts = model.predict(ds_test)

# Remove any single-dimensional entries from the shape (e.g., from (n,1) to (n,))
forecasts = forecasts.squeeze()

# Plot the actual values (X_test) vs. the predicted values (forecasts)
plot_series(time_test, (X_test, forecasts))
```

The results appear promising. Next, we calculate multiple evaluation metrics on the test set and analyze their values relative to the mean of the test data to better understand the model’s performance.

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Convert tensors to NumPy arrays if needed (TensorFlow or PyTorch tensors)
y_true = X_test.numpy() if hasattr(X_test, "numpy") else X_test
y_pred = forecasts.numpy() if hasattr(forecasts, "numpy") else forecasts

# Mean Absolute Error (MAE): average absolute difference between actual and predicted values
mae = mean_absolute_error(y_true, y_pred)

# Mean Squared Error (MSE): average squared difference between actual and predicted values
mse = mean_squared_error(y_true, y_pred)

# Root Mean Squared Error (RMSE): square root of MSE, interpretable in original units
rmse = np.sqrt(mse)

# Mean Absolute Percentage Error (MAPE): average absolute percentage error
mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# Mean Absolute Scaled Error (MASE): scales MAE against a naive forecast
# Naive forecast: previous time step (y_t-1)
naive_forecast = y_true[1:]
naive_actual = y_true[:-1]
mae_naive = mean_absolute_error(naive_actual, naive_forecast)
mase = mae / mae_naive

# R-squared (R²): proportion of variance explained by the model
r2 = r2_score(y_true, y_pred)

# Print evaluation metrics
print(f"MAE   : {mae:.4f}")
print(f"MSE   : {mse:.4f}")
print(f"RMSE  : {rmse:.4f}")
print(f"MAPE  : {mape:.2f}%")
print(f"MASE  : {mase:.4f}")
print(f"R²    : {r2:.4f}%")
```

```python
# Name of the model being evaluated
model_name = "LSTM"

# Create a DataFrame to store the evaluation metrics of the model
# This makes it easy to compare multiple models later
model_results = pd.DataFrame([{
     # Model name
    "model": model_name, 
    # Mean Absolute Error
    "mae": mae,          
     # Mean Squared Error
    "mse": mse, 
    # Root Mean Squared Error
    "rmse": rmse,
     # Mean Absolute Percentage Error
    "mape": mape,  
     # Mean Absolute Scaled Error
    "mase": mase,    
    # R-squared in percentage
    "r2": r2*100           
}])

# Display the first few rows of the results DataFrame
model_results.head()
```

The LSTM model achieved a Mean Absolute Error (MAE) of 6.01, indicating that, on average, its predictions deviate from the actual prices by approximately 6 units. The Mean Squared Error (MSE) was 84.55, and the Root Mean Squared Error (RMSE) was 9.19, reflecting moderate variance in the prediction errors. The Mean Absolute Percentage Error (MAPE) of 12.15% suggests that the model’s predictions are, on average, within 12% of the true values, while the Mean Absolute Scaled Error (MASE) of 1.0252 indicates good relative accuracy compared to a naïve forecast. Finally, the R-squared (R²) value of 90.56% demonstrates that the model explains a high proportion of the variance in the test data, confirming its effectiveness in capturing the underlying patterns of the vegetable price series.

```python
# Sort the model results DataFrame by the 'mae' column in ascending order
model_results = model_results.sort_values(by="mae")
```

```python
# Sort the model results by MAE and plot as a bar chart
model_results[["mae"]].sort_values(by="mae").plot(figsize=(10, 6), kind="bar", legend=False)

# Add a title to the plot
plt.title("Comparison of Models Based on MAE")

# Label the x-axis; here we have only one model, LSTM
plt.xlabel("LSTM")

# Label the y-axis to indicate the metric being measured
plt.ylabel("Mean Absolute Error (MAE)")
```

For the LSTM model, the MAE is 6.01, which indicates that, on average, the predicted prices deviate from the actual prices by about 6 units. This relatively low MAE suggests that the model produces predictions that are close to the true values, making it reliable for forecasting vegetable prices. Visualizing this metric in a bar chart confirms the model’s accuracy and allows for easy comparison with other models if included.

```python
# Sort the model results by R-squared (r2) and plot as a bar chart
# Only the 'r2' column is plotted
# figsize sets the figure size, kind="bar" creates a bar chart
model_results[["r2"]].sort_values(by="r2").plot(figsize=(10, 6), kind="bar")

plt.title("Comparison of Models Based on Model Accuracy (R²)")
plt.xlabel("LSTM")
plt.ylabel("Model Accuracy")
```

The R² value of the LSTM model is 90.56%, showing that the model explains over 90% of the variance in the test dataset. This high R² indicates that the model captures the underlying trends of the vegetable price series effectively. Plotting this metric provides a clear visual representation of how well the model fits the data and highlights its strong predictive performance.

## Conclusion

In this project, we developed an LSTM-based model to forecast vegetable prices using historical time-series data. The process began with careful data preprocessing, where we filtered the dataset for the selected vegetable, retained only the relevant columns (date and average), converted the date column to datetime format, and set it as the DataFrame index. During training, both loss and Mean Absolute Error (MAE) declined steadily, indicating that the model was effectively learning the patterns in the data. Visualizations of training progress and sliding window predictions helped validate the model before evaluating it on the test set. </br>
On the test set, the LSTM model achieved an MAE of 6.01 and an R-squared (R²) of 90.56%, showing that its predictions are, on average, very close to actual prices and that it captures most of the variability in the data. Additional metrics, including MSE (84.55), RMSE (9.19), MAPE (12.15%), and MASE (1.0252), further confirmed the model’s accuracy and robustness. Overall, this workflow—from data preprocessing to LSTM modeling and evaluation—demonstrates the effectiveness of LSTM networks in time-series forecasting and highlights their suitability for predicting vegetable prices with high precision.
